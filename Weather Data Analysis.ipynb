# Visualization

```python
temp = data["Temp_C"]
dew_point_temp = data["Dew Point Temp_C"]
plt.scatter(temp,dew_point_temp)
plt.title("Temperature Against Dew Point Temperature")
plt.xlabel("Temperature (C)")
plt.ylabel("Dew Point Temperature (C)")
plt.show()
```

```python
temp = data["Temp_C"]
Wind_speed = data["Wind Speed_km/h"]
plt.boxplot([temp,Wind_speed], labels = ["Temperature", "Wind_speed"])
plt.title("Temperature Against Wind_speed")
plt.show()
```

```python
temp = data["Temp_C"]
plt.hist(temp)
plt.title("Temperature Distribution")
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.show()
```

```python
Humidity = data["Rel Hum_%"]
plt.hist(Humidity)
plt.title("Humidity Distribution")
plt.xlabel("Humidity (%)")
plt.ylabel("Frequency")
plt.show()
```
